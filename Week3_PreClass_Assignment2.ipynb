## Problem 5

Skewness is a measure of symmetry, or more precisely, the lack of symmetry while Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution. That is, data sets with high kurtosis tend to have heavy tails, or outliers.
In addition, Kurtosis takes a positive value when the distribution is concentrated near the mean from the normal distribution, and takes a negative value when the distribution is scattered from the normal distribution. On the other hand,skewness takes a positive value if distribution is biased to the left and negative value if distribution is biased to the right.

## Problem 6

```python
import seaborn as sns
sns.distplot(df['SalePrice'])
```

```python
var = 'SalePrice'
print("Skewness :", df[var].skew())
print("Kurtosis :", df[var].kurt())
```

```python
log_target = df[var].apply(np.log)
sns.distplot(log_target)
print("Skewness :", log_target.skew())
print("Kurtosis :", log_target.kurt())
```

Before the logarithmic transformation, the kurtosis and skewness were large, and the distribution was likely to follow the Poisson distribution, but after that, the distribution changed to a normal one (normal distribution).
