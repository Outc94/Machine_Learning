## Problem 1

```python
import numpy as np
sigma  =[[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
X = np.random.multivariate_normal(mean=[-3.0,0],cov=sigma,size=500)
```

## Problem 2

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

## Problem 3

```python
plt.figure(1,figsize=(6,10))
plt.subplots_adjust(hspace=0.3)
plt.subplot(2,1,1)
plt.hist(X[:,0],bins=50)
x_min,x_max = np.min(X[:,0]),np.max(X[:,0])
plt.xlim([-6,3])

plt.title('histogram of x1')
plt.xlabel('x1')
plt.ylabel('frequency')

plt.subplot(2,1,2)
plt.hist(X[:,1],bins=50)
plt.xlim([-6,3])
plt.title('histogram of x2')
plt.xlabel('x2')
plt.ylabel('frequency')
```

## Problem 4

```python
sigma  =[[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
Z = np.random.multivariate_normal(mean=[0,-3],cov=sigma,size=500)
plt.scatter(X[:,0],X[:,1],label='0')
plt.scatter(Z[:,0],Z[:,1],color='orange',label='1')
plt.legend()
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
```

## Problem 5

```python
X = np.concatenate([X,Z],axis=0)
X.shape
```
